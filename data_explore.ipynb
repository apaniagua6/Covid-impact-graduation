## 1. 2005 - 2015 Graduation Rates Public School in New York

Graduation rates for New York public schools in 2005 - 2015

```python
import os
import glob
import pandas as pd

grad_rates_new_york = pd.read_csv("data/Graduation Rates Public - School 2005-2015 New York/2005-2015_Graduation_Rates_Public_-_School.csv")
grad_rates_new_york.head()
```

```python
cohort_categories_new_york = grad_rates_new_york['Cohort Category'].unique()
school_names_new_york = grad_rates_new_york['School'].unique()
print("Categories: ", cohort_categories_new_york)
```

```python
print("school_names:", school_names_new_york[:20]) # returns the top 20 from the ndarray
```

## 2. 2014 - 2019 Graduation Rates Public School in Washington

Graduation rates for Washington public schools in 2014 - 2019


```python
# Load Washington State Data Frames for 2014-2019 Disaggregated Data

datapath_washington = 'data/washington/disaggregated_demographics_2014-2019'
filenames_washington = glob.glob( os.path.join(datapath_washington, "*.csv"))

dfs_washington = (pd.read_csv(files, low_memory=False) for files in filenames)
grad_rates_washington = pd.concat(dfs_washington, ignore_index=True, sort=True)
grad_rates_washington.head()
```
